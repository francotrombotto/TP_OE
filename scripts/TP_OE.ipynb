```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Leer el archivo y convertir la columna de fechas
df = pd.read_csv('/sales_sample_2024.csv')
df['sales_date'] = pd.to_datetime(df['sales_date'])

# 2. Crear una nueva columna solo con el Mes (agrupación)
df['mes'] = df['sales_date'].dt.strftime('%Y-%m')

# 3. Sumar el total de ventas por cada mes
ventas_mensuales = df.groupby('mes')['sales_amount'].sum().reset_index()

# 4. Configurar el tamaño y estilo del gráfico
plt.figure(figsize=(10, 6))

# 5. Crear un gráfico de barras
plt.bar(ventas_mensuales['mes'], ventas_mensuales['sales_amount'], color='cornflowerblue', edgecolor='black')

# 6. Agregar títulos, etiquetas y diseño profesional
plt.title('Total de Ventas Mensuales en 2024', fontsize=14, fontweight='bold')
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Monto Total Vendido', fontsize=12)
plt.xticks(rotation=45) # Gira los nombres de los meses para que no se superpongan
plt.grid(axis='y', linestyle='--', alpha=0.7) # Agrega líneas horizontales suaves
plt.tight_layout() # Ajusta los márgenes automáticamente

# 7. Guardar el gráfico mejorado en la carpeta de resultados
plt.savefig('grafico_ventas_mensuales.png')

# 8. Mostrar el gráfico en Colab
plt.show()
```
